## Cleaning of NewCustomerList 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xlrd
%matplotlib inline
```

```python
NewCustm = pd.read_excel('NewCustomerList_table.xlsx')
```

```python
NewCustm.head()
```

```python
NewCustm.isnull().sum()
```

```python
NewCustm['last_name']=NewCustm['last_name'].fillna(method='ffill')
```

```python
NewCustm['last_name'].value_counts()
```

```python
NewCustm['DOB']=NewCustm['DOB'].fillna(method='ffill')
```

```python
NewCustm['job_title'].value_counts()
```

```python
NewCustm['job_title']=NewCustm['job_title'].fillna(method='bfill')
```

```python
NewCustm['job_title'].value_counts()
```

```python
NewCustm['job_industry_category'].value_counts()
```

```python
NewCustm['job_industry_category']=NewCustm['job_industry_category'].fillna(method='bfill')
```

```python
NewCustm['job_industry_category'].value_counts()
```

```python
final_NewCustomerList = pd.DataFrame({
        "first_name": NewCustm["first_name"],
        "last_name": NewCustm["last_name"],
        "gender": NewCustm["gender"],
        "past_3_years_bike_related_purchases": NewCustm["past_3_years_bike_related_purchases"],
        "DOB": NewCustm["DOB"],
        "job_title": NewCustm["job_title"],
        "job_industry_category": NewCustm["job_industry_category"],
        "wealth_segment": NewCustm["wealth_segment"],
        "deceased_indicator": NewCustm["deceased_indicator"],
        "owns_car": NewCustm["owns_car"],
        "tenure": NewCustm["tenure"],
        "address": NewCustm["address"],
        "postcode": NewCustm["postcode"],
        "state": NewCustm["state"],
        "country": NewCustm["country"],
        "property_valuation": NewCustm["property_valuation"],
        "Rank": NewCustm["Rank"],
        "Value": NewCustm["Value"]
    })
```

```python
final_NewCustomerList.to_csv('./final_NewCustomerList.csv', index=False)
```
